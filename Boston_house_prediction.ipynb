```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
```

```python
#independent and dependent features
x=dataset[['PTRATIO','LSTAT']]
y=df.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10) #20% for testing
#random state selects any 80% from the training data, more the random_state greater randomly it
#will select
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
predictions=lr.predict(x_test)
```

```python
from sklearn import metrics
```

```python
print("accuracy score: ",metrics.r2_score(y_test,predictions)) #model performance for testing set
```

```python
p2=lr.predict(x_train)
#model performance for training set
print("accuracy score: ",metrics.r2_score(y_train,p2)) #model performance for training set
```
